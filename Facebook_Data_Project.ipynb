```python
name_outiers = ['Page total likes' ,'Post Month' , 'Post Weekday' , 'Post Hour' ,
                'Paid' , 'Type_Link' , 'Type_Photo' , 'Type_Status' , 'Type_Video' ,
                'Lifetime Post Total Reach' , 'Lifetime Post Total Impressions' ,
                'Lifetime Engaged Users' , 'Lifetime Post Impressions by people who have liked your Page' ,
                'Lifetime Post reach by people who like your Page' ,
                'Lifetime People who have liked your Page and engaged with your post' ,
                'comment' , 'like' , 'share ' , 'Total Interactions'
                ]

```

```python
#for i in name_outiers:
# Q1 = data[i].quantile(0.25)
# Q3 = data[i].quantile(0.75)
# IQR = Q3 - Q1
# upper_limit = Q3 + 1.5 * IQR
# lower_limit = Q1 - 1.5 * IQR
# data.loc[((data[i]>upper_limit , i ))] =upper_limit
#  data.loc[((data[i]<lower_limit , i ))] =lower_limit
```

```python
for i in name_outiers:
  if data[i].dtype != bool:
    Q1 = data[i].quantile(0.25)
    Q3 = data[i].quantile(0.75)
    IQR = Q3 - Q1
    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR
    data.loc[((data[i]>upper_limit) , i )] =upper_limit
    data.loc[((data[i]<lower_limit) , i )] =lower_limit
  else:
    print(f"Skipping outlier handling for boolean column: {i}")
```
